```python
import ddf.minim.*;
import ddf.minim.ugens.*;

Minim minim;
AudioOutput out;

// Signals
Oscil sineOsc;
Noise whiteNoise;
Noise pinkNoise;
Noise brownNoise;

// State control
String mode = "SINE";

void setup() {
  size(500, 300);

  minim = new Minim(this);
  out = minim.getLineOut();

  // --- SINE WAVE ---
  sineOsc = new Oscil(440, 0.5f, Waves.SINE);

  // --- NOISE TYPES ---
  whiteNoise = new Noise(Noise.Tint.WHITE);
  pinkNoise  = new Noise(Noise.Tint.PINK);
  brownNoise  = new Noise(Noise.Tint.BROWN);

  // Start with sine
  sineOsc.patch(out);
}

void draw() {
  background(20);

  fill(255);
  textSize(18);
  text("Audio Signal Generator", 20, 30);

  textSize(14);
  text("Current Mode: " + mode, 20, 60);

  // --- Buttons ---
  drawButton("SINE", 20, 100);
  drawButton("WHITE NOISE", 20, 150);
  drawButton("PINK NOISE", 200, 100);
  drawButton("BROWN NOISE", 200, 150);

  text("Click buttons to switch signals", 20, 250);
}

void drawButton(String label, int x, int y) {
  fill(60);
  rect(x, y, 160, 40, 10);

  fill(255);
  text(label, x + 10, y + 25);
}

void mousePressed() {

  // SINE
  if (mouseOver(20, 100)) {
    switchToSine();
  }

  // WHITE
  if (mouseOver(20, 150)) {
    switchToWhite();
  }

  // PINK
  if (mouseOver(200, 100)) {
    switchToPink();
  }

  // BROWN
  if (mouseOver(200, 150)) {
    switchToBrown();
  }
}

boolean mouseOver(int x, int y) {
  return mouseX > x && mouseX < x + 160 &&
         mouseY > y && mouseY < y + 40;
}

// ---------------- SWITCH FUNCTIONS ----------------

void stopAll() {
  sineOsc.unpatch(out);
  whiteNoise.unpatch(out);
  pinkNoise.unpatch(out);
  brownNoise.unpatch(out);
}

void switchToSine() {
  stopAll();
  sineOsc.patch(out);
  mode = "SINE WAVE";
}

void switchToWhite() {
  stopAll();
  whiteNoise.patch(out);
  mode = "WHITE NOISE";
}

void switchToPink() {
  stopAll();
  pinkNoise.patch(out);
  mode = "PINK NOISE";
}

void switchToBrown() {
  stopAll();
  brownNoise.patch(out);
  mode = "BROWN NOISE";
}
```

### Conclusion

So, in this activity, I've created an **Audio Signal Generator** using the Processing environment with the `Minim` library. I've implemented different types of audio signals:

*   **Sine Wave**: A basic, pure tone.
*   **White Noise**: Random noise with equal intensity at different frequencies, often described as 'hissing'.
*   **Pink Noise**: Noise whose power spectral density is inversely proportional to the frequency, sounding 'flatter' than white noise.
*   **Brown Noise**: Noise whose power spectral density is inversely proportional to the frequency squared, sounding 'deeper' than pink noise.

I've set up a simple interactive interface with buttons to switch between these different sounds. This project demonstrates how to generate and control various audio signals programmatically, allowing for a hands-on exploration of sound synthesis!
